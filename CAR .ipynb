```python
import statsmodels.api as sm

x3train_sc_c = sm.add_constant(x3train_sc)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
from sklearn.metrics import *
```

```python
s= SGDRegressor()
s.fit(x3train_sc,y3train)
```

```python
train_pred = s.predict(x3train_sc)
test_pred = s.predict(x3test_sc)
```

```python
mse1 = mean_squared_error(y3train, train_pred)
rmse1 =  np.sqrt(mean_squared_error(y3train, train_pred))
mae1 = mean_absolute_error(y3train, train_pred)
mape1 = mean_absolute_percentage_error(y3train, train_pred)

print('MSE: ', mse1)
print('RMSE: ',rmse1)
print('MAE: ', mae1)
print('MAPE: ', mape1)
```

```python
mse2 = mean_squared_error(y3test, test_pred)
rmse2 = np.sqrt(mean_squared_error(y3test, test_pred))
mae2 = mean_absolute_error(y3test, test_pred)
mape2 = mean_absolute_percentage_error(y3test, test_pred)

print('MSE: ', mse2)
print('RMSE: ', rmse2)
print('MAE: ', mae2)
print('MAPE: ', mape2)
```

```python
rsq1 = r2_score(y3train, train_pred)
rsq1
```

```python
rsq2 = r2_score(y3test, test_pred)
rsq2
```

```python
## # adjusted R-squared
1 - (1-r2_score(y, y_pred)) * (len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
adj_rsq1 = 1 - (1-r2_score(y3train, train_pred)) * (len(y3train)-1)/(len(y3train)-x3train.shape[1]-1)
adj_rsq1
```

```python
# adjusted R-squared
adj_rsq2 = 1 - (1-r2_score(y3test, test_pred)) * (len(y3test)-1)/(len(y3test)-x3test.shape[1]-1)
adj_rsq2
```

```python
sc = pd.DataFrame({'Train': [mse1, rmse1, mae1, mape1, rsq1, adj_rsq1], 
                   'Test': [mse2, rmse2, mae1, mape2, rsq2, adj_rsq2]}, 
                  index=['MSE', 'RMSE', 'MAE', 'MAPE', 'RSquared', 'Adjusted RSquared'])
```
